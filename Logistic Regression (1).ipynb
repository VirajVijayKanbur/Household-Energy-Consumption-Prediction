```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('energy_data.csv')
print(data.columns)
```

```python
data.dropna()
```

```python
data = data.drop(columns=['Date', 'Time', 'Global_active_power', 'Global_reactive_power', 'Voltage','Sub_metering_1','Sub_metering_2','Sub_metering_3'])
```

```python
data.dropna()
```

```python
HIGH = data[data['Class'] == 1]
LOW = data[data['Class'] == 0]
outlier_fraction = len(HIGH)/float(len(LOW))

#use float so python does not round since value is so small
print('Outlier fraction = ',outlier_fraction)
print('Number of instances where Global_intensity is greater than 23.0: {}'.format(len(data[data['Class'] == 1])))
print('Number of instances where Global_intensity is less than or equal to 23.0: {}'.format(len(data[data['Class'] == 0])))
```

```python
from sklearn import datasets, linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split 
X = data.iloc[:,:-1] # column other than "Class", thus X holds values of Global_Intensity
y = data['Class'] # Y holds values of Class
print("X and y sizes, respectively:", len(X), len(y))
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35)
print("Train and test sizes, respectively:", len(X_train), len(y_train), "|", len(X_test), len(y_test))
```

```python
logistic = linear_model.LogisticRegression(C=1e5) 
logistic.fit(X_train, y_train)
print("Score: ", logistic.score(X_test, y_test))
print("Accuracy: ", logistic.score(X_test, y_test)*100)
```
